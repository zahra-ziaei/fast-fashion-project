```python
entrance_May=pd.read_csv('Entrance_0021ac0419170046_Export_Apr-30-2022-to-May-30-2022_3322.csv')
entrance_Apr=pd.read_csv('Entrance_0021ac0419170046_Export_Mar-31-2022-to-Apr-30-2022.csv')
entrance_March=pd.read_csv('Entrance_0021ac0419170046_Export_Feb-28-2022-to-Mar-31-2022.csv')
entrance_Feb=pd.read_csv('Entrance_0021ac0419170046_Export_Jan-31-2022-to-Feb-28-2022.csv')
```

```python
entrance = pd.concat([entrance_May,entrance_Apr,entrance_March,entrance_Feb])
```

```python
entrance.tail()
```

```python
entrance['date'] = pd.to_datetime(entrance['Timestamp(America/New_York)'])
```

```python
entrance['month']= entrance['date'].dt.month
entrance['day']= entrance['date'].dt.day
entrance['hour']= entrance['date'].dt.hour
entrance['minute']= entrance['date'].dt.minute
entrance['day in week']= entrance['date'].dt.dayofweek
```

```python
entrance = entrance[entrance['date'] >= '2022-02-17']
entrance.tail()
```

```python
daily_visit=entrance.loc[(entrance.minute<5)&(entrance.hour==4) ]
```

```python
daily_visit.drop_duplicates(keep='first',subset=['month','day'],inplace=True)
```

```python
daily_visit["Out_lag_1"] = daily_visit["Out"].shift(1)
daily_visit["In_lag_1"] = daily_visit["In"].shift(1)
daily_visit["Entrance L1_lag_1"] = daily_visit['Entrance L1'].shift(1)
daily_visit["Entrance L2_lag_1"] = daily_visit["Entrance L2"].shift(1)
daily_visit["Entrance L3_lag_1"] = daily_visit["Entrance L3"].shift(1)
daily_visit["Entrance L4_lag_1"] = daily_visit["Entrance L4"].shift(1)
daily_visit["Entrance L5_lag_1"] = daily_visit["Entrance L5"].shift(1)
daily_visit["Entrance L6_lag_1"] = daily_visit["Entrance L6"].shift(1)
```

```python
daily_visit['daily_Out']=daily_visit["Out_lag_1"] -daily_visit["Out"]
daily_visit['daily_In']=daily_visit["In_lag_1"] - daily_visit["In"]
daily_visit['daily_L1']=daily_visit["Entrance L1_lag_1"]- daily_visit['Entrance L1']
daily_visit['daily_L2']=daily_visit["Entrance L2_lag_1"] - daily_visit["Entrance L2"]
daily_visit['daily_L3']=daily_visit["Entrance L3_lag_1"] - daily_visit["Entrance L3"]
daily_visit['daily_L4']=daily_visit["Entrance L4_lag_1"] - daily_visit["Entrance L4"]
daily_visit['daily_L5']=daily_visit["Entrance L5_lag_1"] - daily_visit["Entrance L5"]
daily_visit['daily_L6']=daily_visit["Entrance L6_lag_1"] - daily_visit["Entrance L6"]
```

```python
daily_visit.drop(['In', 'Out', 'Entrance L1',  'Entrance L2', 'Entrance L3', 'Entrance L4', 'Entrance L5', 'Entrance L6','Out_lag_1','In_lag_1',
'Entrance L1_lag_1', 'Entrance L2_lag_1', 'Entrance L3_lag_1', 'Entrance L4_lag_1', 'Entrance L5_lag_1', 'Entrance L6_lag_1','Timestamp(America/New_York)'], axis=1, inplace=True)
```

```python
daily_visit=daily_visit.reset_index(drop=True)
```
